```python
import torch
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt

model_id = "CompVis/stable-diffusion-v1-4"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)
pipe = pipe.to("cuda")

# Варіант 2: Футбол у стилі кіберпанк (можна розкоментувати нижче)
prompt = "Cyberpunk football match, robot soccer players, neon cityscape in the background, rain, cinematic action shot, highly detailed"

image = pipe(prompt).images[0]

# 4. Відображення
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis("off")
plt.show()
```
